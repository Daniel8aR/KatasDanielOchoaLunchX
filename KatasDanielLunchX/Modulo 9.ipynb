```python
def averGen(*fuels):
    sumT = (sum(fuels))
    average = sumT/3
    return average

def fuelReport(fuel1,fuel2,fuel3):
    aver = averGen(fuel1,fuel2,fuel3)
    report = f"""Reporte de combustible:  
    Promedio del combustible: {aver}L
    Combustible 1: {fuel1}L
    Combustible 2: {fuel2}L
    Combustible 3: {fuel3}L
    """
    return report

print(fuelReport(50,60,70))
```

```python
def rocketReport(lunchTime,flyTime,dest,intTank,extTank):
    totalTime = lunchTime + flyTime
    totalFuel = intTank + extTank
    report = f"""Reporte del Cohete:
    Destino: {dest}
    Tiempo total de vuelo: {totalTime} hrs
    Combustible total: {totalFuel} L
    """
    return report

print(rocketReport(3,6,"Marte",30,50))

def bestRocketReport(dest,*timeHours,**tanksFuel):
    totalTime = sum(timeHours)
    totalFuel = sum(tanksFuel.values())
    report = f"""Reporte del Cohete:
    Destino: {dest}
    Tiempo total de vuelo: {totalTime} hrs
    Combustible total: {totalFuel} L
    """
    for title,value in tanksFuel.items():
        report += (f'{title}: {value} \n')
    return report

print(bestRocketReport("Marte",2,10,main=200,ext=500))
```
